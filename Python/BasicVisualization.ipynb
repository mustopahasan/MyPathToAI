# Basic Ploting using Matplotlib

```
import matplotlib.pyplot as plt
```

## Basic Plot

```
y = [ i**3 for i in range(10)]

plt.plot(y)
plt.title('Basic Plot')
plt.show()
```

```
plt.figure()
plt.plot(y, color='g')
plt.title(' Stock Price')
plt.xlabel('Time')
plt.ylabel('Price')
```

```
stockA = y
stockB = [i**2 for i in range(10)]
```

```
plt.figure(figsize=(8,6))
plt.plot(stockA, 'x', color='r', label='Stock A')
plt.plot(stockB,'o', color = 'g',label='Stock B')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Stock Price')
plt.legend()
```

```
# another way to use legend
plt.figure()
plt.plot(stockA, color='r')
plt.plot(stockB, color='g')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Stock Price')
plt.legend(['Stock A','Stock B'])
plt.gca().fill_between(range(len(stockA)),stockA,stockB)
```

##Scatter Plot

```
import numpy as np
point =  np.array([ np.random.randn() for i in range(30)])
step = np.array([ i for i in range(30)])
x = 3*step + 10*point
y = -3*step + 10*point +100
```

```
plt.figure()
plt.scatter(step,x,color='r')
plt.scatter(step,y,color='g')
plt.xlabel('Step')
plt.ylabel('Variable')
plt.title('Relation betwen variables and step')
plt.legend(['Red','Blue'],loc='lower center')
```

## Bar Chart

```
x = ['A','B','C','D']
y = [80, 100, 30, 60]
plt.bar(x,y)
```

```
x = [1,4,7,10]
y = [80, 100, 30, 60]
x2 = np.array(x)+0.75
y2 = [60, 90, 20, 70]
plt.bar(x,y, width=0.75)
plt.bar(x2,y2, width=0.75)
```
